```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
# WARNING, FILE "delay_clean.csv" is 2.5 GB  --  added to gitignore

clean_data = pd.read_csv("./Resources/delay_clean.csv", encoding = 'utf-8')
clean = pd.DataFrame(clean_data)
```

## Support Vector Machines Model

```python
clean.sample(10)
```

```python
# Remove "DEP_DEL15" target (delayed more than 15 minutes) from features data

y = clean.DEP_DEL15
X = clean.drop("DEP_DEL15", axis = 1)
```

```python
# Split training/test datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 50, stratify = y)
```

```python
# Create a StandardScaler instance
scaler = StandardScaler()
```

```python
# Fit the StandardScaler
X_scaler = scaler.fit(X_train)
```

```python
# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Instantiate the model
SVM_model = SVC(kernel = 'linear')
```

```python
SVM_model.fit(X_train_scaled, y_train)
```

```python
# Predict with the model
y_pred = SVM_model.predict(X_test_scaled)
results = pd.DataFrame({"Prediction": y_pred, "Actual": y_test
}).reset_index(drop = True)
results.head()
```

```python
# Assess the model
accuracy_score(y_test, y_pred)
```
